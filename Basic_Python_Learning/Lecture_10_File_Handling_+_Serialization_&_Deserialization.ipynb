# Using Context Manager (With)

- it's a good idea to close a file after usage as it will fee up the resources

- if we dont close it, garbage collector would close it

- with keyword closes the file as soon as the usage is over

```python
# with
with open('/content/sample1.txt','w') as f:
  f.write('selmon bhai')
```

```python
f.write('hello')
```

```python
# try f.read() now
with open('/content/sample.txt','r') as f:
  print(f.readline())
```

```python
# moving within a file -> 10 char then 10 char
with open('/content/sample.txt','r') as f:
  print(f.read(10))
  print(f.read(10))
  print(f.read(10))
  print(f.read(10))
#
```

```python
# benefit ? -> to load a big file in memory
big_L = ['hello world' for i in range(1000)]

with open('big.txt','w') as f:
  f.writelines(big_L)
```

```python
with open('big.txt','r') as f:
  chunk_size = 10
  while len(f.read(chunk_size)) > 0:
    print(f.read(chunk_size),end='***')
    f.read(chunk_size)

```

```python
# seek and tell function
with open ('sample.txt','r') as f:
  f.seek(15)
  print(f.read(10))
  print(f.tell())

  print(f.read(10))
  print(f.tell())
```

```python
# seek during write
with open('sample.txt', 'w') as f:
  f.write('hello')
  f.seek(0)
  f.write('Xa')
```

# Serialization and Deserialization

- Serialization - process of converting python data types to JSON format

- Deserialization - process of converting JSON to python data types

what is JSON
image.png

```python
# serialization using json module
# list

import json

L = [1,2,3,4]

with open('demo.json','w') as f:
  json.dump(L,f)
```

```python
# dict
d = {
    'name': 'saif'
    'age': 25,
    'gender':'male'
}

with open('demo.json','w') as f:
  json.dump(d,f,indent=4)
```

```python
# deserialization
import json

with open('demo.json','r') as f:
  d = json.load(f)
  print(d)
  print(type(d))
```

```python
# serialize and deserialize tuple
import json

t = (1,2,3,4,5)

with open('demo.json', 'w') as f:
  json.dump(t,f)
```

```python
# serialize and deserialize a nested dict

d = {
    'student': 'saif',
    'marks': [23,14,34,44,56]
}

with open('demo.json','w') as f:
  json.dump(d,f)
```

# Serializing and Deserializing custom objects

```python
class Person:

  def __init__(self, fname, lname, age, gender):
    self.fname = fname
    self.lname = lname
    self.age = age
    self.gender = gender

# format to printed in
# ->saif ai > 23 gender -> male
```

```python
person = Person('saif','ali',33,'male')
```

```python
# AS a string
import json

def show_object(person):
  if isinstance(person,Person):
    return '{} {} age -> {} gendor -> {}'.format(person.fname,person.lname,person.age,person.gender)


  with open('demo.json','w') as f:
    json.dump(person,f,default=show_object)
```

```python
# AS a dict
import json

def show_object(person):
  if isinstance(person,Person):
    return {'name':person.fname + ''+ person.lname, 'age':person.age, 'gender':person.gender}

  with open('demo.json','w')as f:
    json.dump(person,f,default = show_object, indent=4)
```

```python
# deserializing
import json

with open('demo.json', 'r')as f:
  d = json.load(f)
  print(d)
  print(type(d))
```
